```python
import math

def e_bul(phi):
    e = 3
    while e < phi:
        if math.gcd(e, phi) == 1:
            return e
        e += 2
    return 65537

DH=int(input("Diffie Helman için 1'i, RSA için 2'yi tuşlayınız:"))
if DH==1:
    print("DİKKAT! Diffie-Hellman çözüm arıyorsun!!")
    p=int(input("p değeri gir:"))
    g=int(input("g değeri gir:"))
    a=int(input("a değeri gir:"))
    b=int(input("b değeri gir:"))
    A=(g**a)%p
    print("A public key:",A)
    B=(g**b)%p
    print("B public key:",B)
    K1=(A**b)%p
    K2=(B**a)%p
    print(f"Bulunan anahtarlar K1: {K1}, K2: {K2}")
    if K1==K2:
        print("Gizli ortak anahtar üretildi")
    else:
        print("Gizli ortak anahtar uyuşmazlığı var!")
elif DH==2:
    print("DİKKAT! RSA çözüm arıyorsun!!")
    p=int(input("p değeri gir:"))
    q=int(input("q değeri gir:"))
    n=q*p
    phi=(p-1)*(q-1)
    print(f"Bulunan n: {n}, phi: {phi}")
    e=int(input("e değeri verilmişse gir, verilmemişse 0 gir ben hesaplarım ;) :"))
    if e!=0:
        e=e
    else:
        e=e_bul(phi)
    k = 1
    while True:
        if (k * phi + 1) % e == 0:
            d = (k * phi + 1) // e
            break
        k += 1
    print(f"e:{e},d:{d}")
    M=int(input("M değeri gir:"))
    c=(M**e)%n
    print("Şifreli mesaj:",c)
    f=(c**d)%n
    print("Şifresi çözülmüş mesaj:",f)
    if M==f:
        print("Şifreleme işlemi doğrulandı!!!")
    else:
        print("Şifreleme işlemi başarısız!")
else:
    print("Geçerli bir seçenek tuşlamadınız!!")
```
